```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Sales': [1000, 2500, 3500, 4700, 5800, 6400, 7200, 8500, 9100, 9600, 9800, 10000],
    'Expenses': [1200, 2300, 3200, 4100, 5000, 5600, 6100, 7300, 8100, 8800, 9400, 9900]
}
df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(10, 5))
plt.bar(df['Month'], df['Sales'], color='blue', label='Sales')
plt.xlabel('Month')
plt.ylabel('Sales (in $)')
plt.title('Monthly Sales')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['Month'], df['Sales'], marker='o', linestyle='-', color='blue', label='Sales')
plt.plot(df['Month'], df['Expenses'], marker='o', linestyle='-', color='red', label='Expenses')
plt.xlabel('Month')
plt.ylabel('Amount (in $)')
plt.title('Monthly Sales and Expenses')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(df['Month'], df['Sales'], color='blue', label='Sales')
plt.scatter(df['Month'], df['Expenses'], color='red', label='Expenses')
plt.xlabel('Month')
plt.ylabel('Amount (in $)')
plt.title('Monthly Sales and Expenses')
plt.legend()
plt.show()
```
